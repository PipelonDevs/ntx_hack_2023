```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
boring_game_results = np.genfromtxt(open("../../Datasets/modelTest2_result.csv", "rb"), delimiter=" ")[:16500]
immerse_game_results = np.genfromtxt(open("../../Datasets/modelTest3_result.csv", "rb"), delimiter=" ")[:16500]
```

```python
# plot the results averaged over 10 games
avg_boring_game_results = np.mean(boring_game_results.reshape(-1, 100), axis=1)
avg_immerse_game_results = np.mean(immerse_game_results.reshape(-1, 100), axis=1)

# plot side by side

fig, ax = plt.subplots(2, 1, figsize=(7, 5))
ax[0].set_title("Boring Game")
ax[0].set_ylabel("Average Immerse Score")
# remove x label
ax[0].xaxis.set_visible(False)
ax[0].plot(range(len(avg_boring_game_results)), avg_boring_game_results, label="boring game")

ax[1].set_title("Immersive Game")
ax[1].set_xlabel("Time [4s]")
# set unit to be 1000
ax[1].set_ylabel("Average Immerse Score")
ax[1].plot(range(len(avg_immerse_game_results)), avg_immerse_game_results, label="immerse game")
```

```python
import matplotlib.animation as animation

def update_line(frame):
    l.set_data(range(frame), avg_immerse_game_results[:frame])
    return l

# Animation of Immersive game
fig1, ax = plt.subplots()
l = ax.plot(range(avg_immerse_game_results.shape[0]), avg_immerse_game_results, label="immerse game")[0]

plt.xlabel('Time [4s]')
plt.ylabel('Average Immerse Score')
plt.title('Immersive Game')

line_ani = animation.FuncAnimation(fig1, update_line, interval=60, frames=avg_immerse_game_results.shape[0])
line_ani.save('immersiveLALAL.gif')
```
